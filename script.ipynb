# [dbo].[GetPoslovnica]

```sql
/****** Object:  UserDefinedFunction [dbo].[GetPoslovnica]    Script Date: 20.04.2025. 22:06:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- First create scalar function
CREATE FUNCTION [dbo].[GetPoslovnica](@delivery_note varchar(50))
RETURNS varchar(2)
AS
BEGIN
    DECLARE @poslovnica varchar(2)
    
    SELECT TOP 1 @poslovnica = SUBSTRING(receiver, 3, 2)
    FROM [PY_SHOPS].[dbo].[incoming_goods]
    WHERE delivery_note = @delivery_note 
    AND receiver LIKE 'PJ%'
    
    RETURN @poslovnica
END
GO
```

# [dbo].[v_ProductStock]

```sql
/****** Object:  View [dbo].[v_ProductStock]    Script Date: 20.04.2025. 22:06:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO


CREATE VIEW [dbo].[v_ProductStock] AS
SELECT 
    m.[prd_ID],
    m.[prd_PrimaryCode],
    m.[prd_Description],
    m.[Total Count 2024],
    m.[Total QTY 2024],
    m.[Turnover average],
    m.[QTY average],
    m.[Max Qty],
    m.[Min Qty],
    m.[AttributeName],
    m.[pav_value],
    s.Total_Quantity,
    s.Total_QuantityFree,
    s.[UnitDescription]
FROM [Analize].[dbo].[v_MaxMinQty] m (NOLOCK)
LEFT JOIN (
    SELECT 
        [stk_ProductID],
        SUM([spt_Quantity]) as Total_Quantity,
        SUM([spt_QuantityFree]) as Total_QuantityFree,
        MAX([UnitDescription]) as UnitDescription
    FROM [LVS].[LVision].[dbo].[V_ProductStockSum]
    WHERE [LanguageID] = '1'
    GROUP BY [stk_ProductID]
) s ON m.[prd_ID] = s.[stk_ProductID]
WHERE m.[AttributeName] LIKE '%Warehouse picking profile%'
GO
```

# [dbo].[SP_UpdateIncomingGoodsAnnouncement]

```sql
/****** Object:  StoredProcedure [dbo].[SP_UpdateIncomingGoodsAnnouncement]    Script Date: 20.04.2025. 22:06:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   PROCEDURE [dbo].[SP_UpdateIncomingGoodsAnnouncement]
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Step 1: Get base data into temp table
    SELECT 
        DeliveryNote, SSCC, MSI, ItemCode, 
        ItemDescription, ItemUOM, Quantity,
        Receiver, Customer, DeliveryType, 
        OrderType, OrderCode
    INTO #BaseData
    FROM [LVS].[LVision].[DBO].[v_INSTORE_PackingLists]
    WHERE Receiver LIKE 'PJ%';

    -- Step 2: Aggregate the data
    SELECT 
        DeliveryNote COLLATE Croatian_CI_AS as delivery_note,
        SSCC COLLATE Croatian_CI_AS as sscc,
        MSI COLLATE Croatian_CI_AS as document_msi,
        ItemCode COLLATE Croatian_CI_AS as item_code,
        MAX(ItemDescription) COLLATE Croatian_CI_AS as description,
        MAX(ItemUOM) COLLATE Croatian_CI_AS as uom,
        SUM(Quantity) as quantity,
        MAX(Receiver) COLLATE Croatian_CI_AS as receiver,
        MAX(Customer) COLLATE Croatian_CI_AS as customer,
        MAX(DeliveryType) COLLATE Croatian_CI_AS as delivery_type,
        MAX(OrderType) COLLATE Croatian_CI_AS as order_type,
        MAX(OrderCode) COLLATE Croatian_CI_AS as sales_order,
        ROW_NUMBER() OVER (
            PARTITION BY DeliveryNote, SSCC, ItemCode 
            ORDER BY MSI DESC
        ) as rn
    INTO #FinalData
    FROM #BaseData
    GROUP BY DeliveryNote, SSCC, MSI, ItemCode;

    -- Step 3: Merge the data
    MERGE [dbo].[incoming_goods_announcement] AS target
    USING (SELECT * FROM #FinalData WHERE rn = 1) AS source
    ON (
        target.delivery_note = source.delivery_note
        AND target.sscc = source.sscc
        AND target.item_code = source.item_code
    )
    WHEN MATCHED AND target.document_msi IS NULL 
        AND source.document_msi IS NOT NULL THEN
        UPDATE SET target.document_msi = source.document_msi
    WHEN NOT MATCHED BY TARGET THEN
        INSERT (
            delivery_note, sscc, sales_order, document_msi,
            item_code, description, uom, quantity,
            receiver, customer, delivery_type, order_type,
            barcode
        )
        VALUES (
            source.delivery_note, source.sscc, source.sales_order, source.document_msi,
            source.item_code, source.description, source.uom, source.quantity,
            source.receiver, source.customer, source.delivery_type, source.order_type,
            source.item_code
        );

    -- Step 4: Update PSSCC
    UPDATE iga
    SET iga.PSSCC = ps.PSSCC
    FROM [dbo].[incoming_goods_announcement] iga
    INNER JOIN [LVS].[LVision].[DBO].[v_INSTORE_PackingSSCC] ps 
    ON iga.sscc COLLATE Croatian_CI_AS = ps.SSCC COLLATE Croatian_CI_AS
    WHERE ps.PSSCC LIKE '3502%';

    -- Clean up
    DROP TABLE #BaseData;
    DROP TABLE #FinalData;
END
GO
```

# [dbo].[SP_UpdateMasterData]

```sql
/****** Object:  StoredProcedure [dbo].[SP_UpdateMasterData]    Script Date: 20.04.2025. 22:06:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_UpdateMasterData]
AS
BEGIN
    SET NOCOUNT ON;

    -- Step 1: Create temp table with additional columns
    SELECT DISTINCT
        p.BarCode COLLATE Croatian_CI_AS as barcode,
        p.ItemCode COLLATE Croatian_CI_AS as item_code,
        p.ItemDescription COLLATE Croatian_CI_AS as description,
        p.CU COLLATE Croatian_CI_AS as uom,
        p.cuperPAK as pack
    INTO #TempNewRecords
    FROM [LVS].[LVision].[DBO].[v_INSTORE_Products] p
    WHERE p.BarCode IS NOT NULL;

    -- Step 2: Remove duplicates within temp table
    WITH DuplicatesInTemp AS (
        SELECT barcode
        FROM #TempNewRecords
        GROUP BY barcode
        HAVING COUNT(*) > 1
    )
    DELETE t
    FROM #TempNewRecords t
    WHERE EXISTS (
        SELECT 1 FROM DuplicatesInTemp d
        WHERE d.barcode = t.barcode
    );

    -- Step 3: Update existing records where UOM or pack is empty
    UPDATE m
    SET 
        m.uom = CASE WHEN m.uom IS NULL OR m.uom = '' THEN t.uom ELSE m.uom END,
        m.pack = CASE WHEN m.pack IS NULL THEN t.pack ELSE m.pack END
    FROM [PY_SHOPS].[dbo].[Master_data] m
    INNER JOIN #TempNewRecords t ON m.barcode = t.barcode;

    -- Step 4: Delete existing barcodes from temp table
    DELETE t
    FROM #TempNewRecords t
    WHERE EXISTS (
        SELECT 1 
        FROM [PY_SHOPS].[dbo].[Master_data] m 
        WHERE m.barcode = t.barcode
    );

    -- Step 5: Insert remaining records
    INSERT INTO [PY_SHOPS].[dbo].[Master_data] 
    (barcode, item_code, description, uom, pack)
    SELECT barcode, item_code, description, uom, pack
    FROM #TempNewRecords;

    -- Clean up
    DROP TABLE #TempNewRecords;
END
GO
```
